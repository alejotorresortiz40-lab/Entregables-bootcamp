```python
#Parte 1

# Consultar datos

#Utilizamos la API de MarketStack para obtener datos históricos de precios de acciones.
#En este caso, se consultan los datos de cierre diario de Apple (AAPL) con un límite de 100 registros.
#Se utiliza un timeout de 10 segundos para evitar que la solicitud se quede colgada indefinidamente.

import requests
import csv

response = requests.get(
    "https://api.marketstack.com/v1/eod",
    params={
        "access_key": "6f60f7f4f5512ecd3a090aa1b888a338",
        "symbols": "AAPL",
        "limit": 100
    },
    timeout=10
)

data = response.json()

# Verificar que llegaron datos
if "data" not in data:
    print("Error:", data)
else:
    registros = []
# Transformar y estructurar datos
    for day in data["data"]:
        registros.append({
            "fecha": day["date"][:10],
            "simbolo": day["symbol"],
            "apertura": day["open"],
            "maximo": day["high"],
            "minimo": day["low"],
            "cierre": day["close"],
            "volumen": int(day["volume"]),
            "dividendo": day["dividend"],
            "exchange": day["exchange"]
        })

    # Guardar datos en archivo CSV
    campos = ["fecha"; "simbolo"; "apertura"; "maximo"; "minimo"; "cierre"; "volumen"; "dividendo"; "exchange"]

    with open("aapl_datos.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=campos)
        writer.writeheader()
        writer.writerows(registros)

    print(f" CSV creado con {len(registros)} registros")
    print(f" Archivo: aapl_datos.csv")

    # Vista previa
    for r in registros[:3]:
        print(r)
```

```python
#Transformación de datos porque Power BI no reconoce el formato de los precios del stock de AAPL

#Debido a la limitación en las consultas que se podían hacer, se eligió solo 1 acción con datos de 1 exchange,
#pero se podrían agregar más acciones y exchanges sin problemas.

import requests
import csv

response = requests.get(
    "https://api.marketstack.com/v1/eod",
    params={
        "access_key": "6f60f7f4f5512ecd3a090aa1b888a338",
        "symbols": "AAPL",
        "limit": 100
    },
    timeout=10
)

data = response.json()

registros = []
for day in data["data"]:
    registros.append({
        "fecha": day["date"][:10],
        "simbolo": day["symbol"],
        "apertura": day["open"],
        "maximo": day["high"],
        "minimo": day["low"],
        "cierre": day["close"],
        "volumen": int(day["volume"]),
        "dividendo": day["dividend"],
        "exchange": day["exchange"]
    })

campos = ["fecha", "simbolo", "apertura", "maximo", "minimo", "cierre", "volumen", "dividendo", "exchange"]

with open("aapl_datos.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=campos, delimiter=";")
    writer.writeheader()
    for r in registros:
        writer.writerow({
            "fecha": r["fecha"],
            "simbolo": r["simbolo"],
            "apertura": str(r["apertura"]).replace(".", ","),
            "maximo": str(r["maximo"]).replace(".", ","),
            "minimo": str(r["minimo"]).replace(".", ","),
            "cierre": str(r["cierre"]).replace(".", ","),
            "volumen": r["volumen"],
            "dividendo": str(r["dividendo"]).replace(".", ","),
            "exchange": r["exchange"]
        })

print(f"CSV creado con {len(registros)} registros")
print(f"Archivo: aapl_datos.csv")
```
